#### Ollama
Ollama supports embedding models, making it possible to build retrieval augmented generation (RAG) applications that combine text prompts with existing documents or other data.

In Ollama we are downloading embedding models locally to our own machine. for example by pasting "ollama pull mxbai-embed-large" on command prompt we are pulling this embedding model into our local PC.

```python
from langchain_community.embeddings import OllamaEmbeddings
```

```python
embeddings=(
    OllamaEmbeddings(model="gemma:2b")  ##by default it ues llama2
)
```

```python
embeddings
```

```python
r1=embeddings.embed_documents(
    [
       "Alpha is the first letter of Greek alphabet",
       "Beta is the second letter of Greek alphabet", 
    ]
)
```

```python
# dimension with respect to vectors = 2048
len(r1[0])
```

```python
r1[1]
```

```python
embeddings.embed_query("What is the second letter of Greek alphabet ")
```

```python
### Other Embedding Models
### https://ollama.com/blog/embedding-models
embeddings = OllamaEmbeddings(model="mxbai-embed-large")
text = "This is a test document."
query_result = embeddings.embed_query(text)
query_result
```
