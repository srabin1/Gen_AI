```python
import nltk
nltk.download('punkt_tab')
```

```python
corpus="""Hello Welcome,to Krish Naik's NLP Tutorials.
Please do watch the entire course! to become expert in NLP.
"""
```

```python
print(corpus)
```

### 1. Tokenization: convert paragraphs --> sentences

```python
from nltk.tokenize import sent_tokenize
```

```python
sent_tokenize(corpus)
```

```python
documents = sent_tokenize(corpus)
```

```python
type(documents)
```

```python
for sentence in documents:
  print(sentence)
```

### 2. Tokenization: convert paragraphs --> words

```python
from nltk import word_tokenize
```

```python
word_tokenize(corpus)
```

```python
for sentence in documents:
  print(word_tokenize(sentence))
```

Below library will recognize `punctuations`. in previous code `'s` was considered as one element but with `wordpunct_tokenize` they are two different elements `s` and `'`.

```python
from nltk import wordpunct_tokenize
```

```python
wordpunct_tokenize(corpus)
```

in `TreebankWordTokenizer` only last full stop will be considered as a separator, otherwise the rest of `.` will be merged with its previous word and is considered as one item.

```python
from nltk import TreebankWordTokenizer
```

```python
tokenizer = TreebankWordTokenizer()
```

```python
tokenizer.tokenize(corpus)
```
