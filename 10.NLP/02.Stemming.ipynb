### Stemming is the process of reducing a word to its word stem that affixes to suffixes and prefixes or to the roots of words knwon as a lemma.

## Classification problem
We want to investigate if the comments on the product is a positive review or a negative review.

Reviews can be like: eating, eat, eaten or [going, gone, goes]--> go (is the stem root)

```python
words=["eating","eats","eaten","writing","writes","programming","programs","history","finally","finalized"]
```

## 1. The first stemming technique will be porter stemmer

```python
from nltk.stem import PorterStemmer
```

```python
stemming = PorterStemmer()
```

```python
for word in words:
  print(word+"--->"+stemming.stem(word))
```

The drawback of stemming is changing the meaning of the root of some words like: history --> histori

```python
stemming.stem('congratulatoins')
```

## 2. Regular expression stemmer

```python
from nltk.stem import RegexpStemmer
```

RegexpStemmer($ing$|$s$|$e$|$able$, min=4)

This means wherever the "last" part in a word is either ing, s, e, or able remove that.

```python
reg_stemmer=RegexpStemmer('ing$|s$|e$|able$', min=4)
```

```python
reg_stemmer.stem('eating')
```

```python
reg_stemmer.stem('ingeating')
```

## 3. Snowball stemmer

It is more accurate and powerfull compared with Porter stemmer

```python
from nltk.stem import SnowballStemmer
```

```python
snowball_stemmer = SnowballStemmer('english')
```

```python
for word in words:
  print(word + "--->" + snowball_stemmer.stem(word))
```

Difference between Porter stemming and Snowball stemming

```python
# with porter stemming
stemming.stem("fairly"), stemming.stem("sportingly")
```

```python
# with snowball stemming
snowball_stemmer.stem("fairly"), snowball_stemmer.stem("sportingly")
```

```python
snowball_stemmer.stem("goes")
```
