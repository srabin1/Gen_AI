### We can resolve the issue of stemming which could find the wrong root with lemmatization

Use case: Q&A, chatbots, text summarization

lemmatizer will give us the exact root word

```python
!pip install nltk
```

```python
import nltk
nltk.download('wordnet')
```

lemmatizer pos:\
Noun: n, Verb: v, Adjective: a, Adverb: r

```python
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
lemmatizer.lemmatize("going")
```

```python
lemmatizer.lemmatize("going", pos = 'v')
```

```python
lemmatizer.lemmatize("going", pos = 'n')
```

```python
words=["eating","eats","eaten","writing","writes","programming","programs","history","finally","finalized"]
```

```python
for word in words:
  print(word + "--> " +lemmatizer.lemmatize(word))
```

```python
for word in words:
  print(word + "--> " +lemmatizer.lemmatize(word, pos = 'v'))
```
